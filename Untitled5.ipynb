```python
from google.colab import files
import pandas as pd
import io

print("Por favor, selecione o arquivo CSV da PRF:")
uploaded = files.upload()

nome_arquivo = list(uploaded.keys())[0]
print(f"\nArquivo '{nome_arquivo}' carregado com sucesso! Lendo os dados...")

df = pd.read_csv(io.BytesIO(uploaded[nome_arquivo]), sep=';', encoding='latin-1')

print(f"\nTamanho do dataset: {df.shape[0]} linhas e {df.shape[1]} colunas\n")
display(df.head())
```
